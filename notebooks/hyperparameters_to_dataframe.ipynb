# Create DataFrame for automated training of networks
Below, we define a dataframe in which training hyperparameters of interest for experimentation and more mundane parameters like flags for saving models are listed. Work in progress.


```python
%load_ext autoreload
%autoreload 2
#%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sys, os, time
# path to current project
sys.path.append("/mnt/hd_internal/hh/projects_DS/numeric_precision/")
# path into which to save dataframe
df_dir = "../models"
```

```python
# total number of model runs
num_runs = 10

# model type ("convnet_01", "VGG16" or "InceptionV3")
#parameters_train = pd.DataFrame(np.tile(np.array(["convnet_01"]), [num_runs, 1])  , columns=["model_type"])
#parameters_train = pd.DataFrame(np.tile(np.array(["VGG16"]), [num_runs, 1])  , columns=["model_type"])
parameters_train = pd.DataFrame(np.tile(np.array(["InceptionV3"]), [num_runs, 1])  , columns=["model_type"])

# model weights
parameters_train["model_weights"] = np.tile(np.array(["imagenet"]), [num_runs, 1])
#parameters_train["model_weights"] = np.tile(np.array(["None"]), [num_runs, 1])


# name of model 
parameters_train["model_name"] = np.tile(np.array(["undefined"]), [num_runs, 1])

# batch size 
parameters_train["batch_size"] = np.ones((num_runs, 1), dtype=int)*32

# number of epochs
parameters_train["num_epochs"] = np.ones((num_runs, 1), dtype=int)*200

# perform image augmentation on training set?
parameters_train["do_img_augment"] = np.ones((num_runs, 1), dtype=bool)

# this parameter decides which of the two combinations of last layer activation 
# and implementation of loss function shall be used for training; set to either
# "raw" or "sigmoid"
parameters_train["loss_basis"] = np.tile(np.array([["sigmoid"], ["raw"]]), [num_runs // 2, 1])

# shall tensorboard logging be performed?
parameters_train["do_tensorboard"] = np.ones((num_runs, 1), dtype=bool)

# how to save model, if at all? "ModelCheckpoint" or "None"
parameters_train["save_model"] = np.tile(np.array(["None"]), [num_runs, 1])
parameters_train.loc[:10, "save_model"] = "ModelCheckpoint"

# save dataframe to file
parameters_train.to_csv(os.path.join(df_dir, "train_params.csv"), index=True)

# peek at its contents
parameters_train.head(20)
```
